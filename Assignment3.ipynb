### Question 5

#### Plot 1- Scatter plot - SL vs SW

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SL', y='SW', hue='Classification', palette='Set1', s=100)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Classification')
plt.grid(True)
plt.show()
```

#### Plot 2- Scatter plot - PL vs PW

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PL', y='PW', hue='Classification', palette='Set1', s=100)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Classification')
plt.grid(True)
plt.show()

```

#### Plot 3- Bar graph - SL vs PL

```python
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Bar Graph of SL vs PL')
sns.barplot(data=df,x=df['SL'],y=df['PL'],hue='Classification')
plt.xticks(rotation=90)
plt.figure(figsize=(12,6))
plt.show()
```
